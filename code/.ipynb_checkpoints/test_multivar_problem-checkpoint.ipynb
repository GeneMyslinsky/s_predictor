```python
# some example data
import numpy as np
import pandas
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
# mdata = sm.datasets.macrodata.load_pandas().data
from statsmodels.tsa.base.datetools import dates_from_str

sns.set()

req_cols = ['average','ao','vpt','macd']
data = df[req_cols]
df_h,df_t = data.head(1),data.tail(1)
def makeDiff(data):
    data = data.diff()
    data = data.replace([np.inf, -np.inf], np.nan)
    data.dropna(inplace=True)
    return data

def makeDiffLog(data):
    data = np.log(data).diff()
    data = data.replace([np.inf, -np.inf], np.nan)
    data.dropna(inplace=True)
    return data
# data.to_csv('check.csv')
def makeModel(data):
    model = VAR(data)
    results = model.fit(24)
    results.summary()
    lag_order = results.k_ar
    return model,results,lag_order
def makeForecast(data,nper,lag):
    f = results.forecast(data.values[-lag_order:], nper)
    return f

display(data)
mdata = makeDiffLog(data)
model,results,lag_order = makeModel(mdata)
f = makeForecast(mdata,400,240)

df_prediction_diff = pd.DataFrame(f,columns=req_cols)


df_prediction_test = pd.concat([df_h,df_prediction_diff])
display(df_prediction_test.cumsum().plot.line())
```
